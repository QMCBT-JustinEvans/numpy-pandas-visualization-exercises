2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

* y = âˆšx

* y = x^3

* y = 2^x

* y = 1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
x = (-10, -5, 0, 5, 10)
y1 = [(np.sqrt(np.absolute(x))) for x in x]
y2 = [(x ** 3) for x in x]
y3 = [(np.power(2, np.absolute(x))) for x in x]
y4 = [(1 / (x + 1)) for x in x]

print(x, y1)
print(x, y2)
print(x, y3)
print(x, y4)
```

```python
# plot each mathematical function in a separate figure

plt.figure(figsize = (4,4))

plt.subplot(111)
plt.plot(x, y1,color = 'r', ls = '--')
plt.title('y = âˆšx')

plt.show()
```

```python
# plot each mathematical function in a separate figure

plt.figure(figsize = (4,4))

plt.subplot(111)
plt.plot(x, y2,color = 'r', ls = '--', marker = '^')
plt.title('y = x^3')

plt.show()
```

```python
# plot each mathematical function in a separate figure

plt.figure(figsize = (4,4))

plt.subplot(111)
plt.plot(x, y3,color = 'g', ls = '--', marker = 'o')
plt.title('y = 2^x')

plt.show()
```

```python
# plot each mathematical function in a separate figure

plt.figure(figsize = (4,4))

plt.subplot(111)
plt.plot(x, y4,color = 'm', ls = '--', marker = 'v')
plt.title('y = 1/(x+1)')

plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# plot different mathematical functions vs x in a single figure

plt.figure(figsize = (8,8))
plt.suptitle('Math is Hard')

plt.subplot(221)
plt.plot(x, y1,color = 'r', ls = '--')
plt.title('y = âˆšx')

plt.subplot(222)
plt.plot(x, y2,color = 'r', ls = '--', marker = '^')
plt.title('y = x^3')

plt.subplot(223)
plt.plot(x, y3,color = 'g', ls = '--', marker = 'o')
plt.title('y = 2^x')

plt.subplot(224)
plt.plot(x, y4,color = 'm', ls = '--', marker = 'v')
plt.title('y = 1/(x+1)')

plt.tight_layout()

plt.show()
```
