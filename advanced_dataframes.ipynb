1. Copy the users and roles DataFrames from the examples above.

````python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
````

````python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
````

2. What is the result of using a right join on the DataFrames?

    * ANSWER: All data is kept from the table on the right ("roles" as the second table called with ``` ".merge(roles,"```) all missing data that does not match from the left "users" table will be filled in as "NaN".

````python
users.merge(roles, left_on='role_id', right_on='id', how='right', indicator=True)
````

3. What is the result of using an outer join on the DataFrames?

    * ANSWER: All records from both tables are kept with unmatched data filled as "NaN".

````python
users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
````

4. What happens if you drop the foreign keys from the DataFrames and try to merge them?

    * ANSWER: It will error out if you leave the left_on and right_on as empty quotes ""
              It will mismatch your data if align them both on "id" without using the foriegn key
              For left and right with no "_on" statement, 
                  it will bring back headers for both,
                  it will fill data for only left or right (according to "how"),
                  no matches will be made between the two,
                  and all data from the column not in the primary table being kept (how= L or R)
                  will be filled with "NaN"
              Outer will return both full data tables 
                  with no matched data and all headers not in original table filled with "NaN"
              Inner will bring back only the headers from both tables because there ae no matched records

````python
users.merge(roles, left_on='id', right_on='id', how='outer', indicator=True)
````

````python
users.merge(roles, how='left', indicator=True)
````

````python
users.merge(roles, how='right', indicator=True)
````

````python
users.merge(roles, how='outer', indicator=True)
````

````python
users.merge(roles, how='inner', indicator=True)
````
